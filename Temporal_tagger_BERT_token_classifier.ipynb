```python
from transformers import AutoTokenizer, AutoModelForTokenClassification, pipeline

# O -- outside of a tag
# I-TIME -- inside tag of time
# B-TIME -- beginning tag of time
# I-DATE -- inside tag of date
# B-DATE -- beginning tag of date
# I-DURATION -- inside tag of duration
# B-DURATION -- beginning tag of duration
# I-SET -- inside tag of the set
# B-SET -- beginning tag of the set

tokenizer = AutoTokenizer.from_pretrained("satyaalmasian/temporal_tagger_BERT_tokenclassifier")

model = AutoModelForTokenClassification.from_pretrained("satyaalmasian/temporal_tagger_BERT_tokenclassifier")

nerpipeline = pipeline('ner', model=model, tokenizer=tokenizer)
text = "I'm meeting my friend today at the park. We will meet tomorrow at 4 PM. i wish my exams where next month. but it 3rd this week"
print(nerpipeline(text))
```
